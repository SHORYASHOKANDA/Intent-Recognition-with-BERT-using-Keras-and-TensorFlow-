```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload the image file
uploaded = files.upload()

# Read the uploaded image
file_name = next(iter(uploaded))
img = cv2.imread(file_name)

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Reduce the contrast using the equalizeHist function
equalized_img = cv2.equalizeHist(gray_img)

# Convert the equalized image back to BGR format
result_img = cv2.cvtColor(equalized_img, cv2.COLOR_GRAY2BGR)

# Display the original and contrast-adjusted images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')
axes[1].imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
axes[1].set_title("Contrast-Adjusted Image")
axes[1].axis('off')

plt.show()
```

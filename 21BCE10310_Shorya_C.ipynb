```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# Upload the image file
uploaded = files.upload()

# Read the uploaded image
file_name = next(iter(uploaded))
img = cv2.imread(file_name)

# Flip the image horizontally
flipped_img = np.fliplr(img)

# Convert the image from BGR to RGB (Matplotlib expects RGB format)
flipped_img_rgb = cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB)

# Display the original and flipped images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')
axes[1].imshow(flipped_img_rgb)
axes[1].set_title("Flipped Image (Horizontally)")
axes[1].axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# Upload the image file
uploaded = files.upload()

# Read the uploaded image
file_name = next(iter(uploaded))
img = cv2.imread(file_name)

# Flip the image both vertically and horizontally
flipped_img = np.flip(img, (0, 1))

# Convert the image from BGR to RGB (Matplotlib expects RGB format)
flipped_img_rgb = cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB)

# Display the original and flipped images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')
axes[1].imshow(flipped_img_rgb)
axes[1].set_title("Flipped Image (Vertically and Horizontally)")
axes[1].axis('off')

plt.show()
```
